## Plot a subgraph of these two Long ARG outliers

NB: this uses a number of hardcoded node numbers. This figure is also generated by the plot.py script

```python
np.unique(ts_long.edges_child[ts_long.edges_parent == 232088]).shape
```

```python
node_col = {}

focal_nodes = [229998, 235293]  # The two focal recombination nodes
for u in focal_nodes:
    node_col[u] = "darkred"

to_plot = set()
for u in set(focal_nodes) | {248215, 243149, 204945}: # include children with many descendants
    to_plot.add(u)
    for child in np.unique(ts_long.edges_child[ts_long.edges_parent == u]):
        to_plot.add(child)

# Add the parents of each focal recombination node
for u in focal_nodes:
    to_plot.update(ts_long.edges_parent[ts_long.edges_child == u])

# Add the main Delta polytomy
main_delta_source = 261771
node_col[main_delta_source] = "gold"
to_plot.add(main_delta_source)

def is_Delta(pango):
    return pango.startswith("AY") or pango == "B.1.617.2"

desc_target = collections.defaultdict(set)
for tree in tqdm.tqdm(ts_long.trees(), desc=f"Finding descendants of {main_delta_source} and others"):
    for target in (main_delta_source, 232088):
        for u in tree.samples(target):
            desc_target[target].add(u)

tot_delta = 0
desc_delta = collections.defaultdict(int)
for u in (u for k, v in ti_long.pango_lineage_samples.items() for u in v if is_Delta(k)):
    tot_delta += 1
    for k, v in desc_target.items():
        if u in v:
            desc_delta[k] += 1

desc = desc_delta.pop(main_delta_source)
print(
    f"Node {main_delta_source} (coloured {node_col[main_delta_source]}, Pango",
    f"{ts_long.node(main_delta_source).metadata['Imputed_Nextclade_pango']}) has",
    len(np.unique(ts_long.edges_child[ts_long.edges_parent == main_delta_source])),
    f"children ({desc/tot_delta * 100:.02f}% of Delta samples)"
)

for k, v in desc_delta.items():
    print(f"Node {k} is {v/tot_delta * 100:.02f}% of Delta samples")

# Add path between the two major parents and their MRCA
tree = ts_long.at(21400)
trace_parents = [21605, 204945]
mrca = tree.mrca(*trace_parents)
for u in trace_parents:
    while True:
        u = tree.parent(u)
        if u < 0:
            break
        to_plot.add(u)
        if u == mrca:
            break

node_col[None] = "lightgray"
node_col["Unknown (R)"] = "k"
node_col["Unknown"] = "white"
node_col["B.1.617.2"] = "#e2da91"
node_col["AY.100"] = "#e2da91"
node_col["AY.112"] = "#e2da91"

from matplotlib import pyplot as plt
fig, ax = plt.subplots(1,1,figsize=(25, 30))

params = dict(
    ts_id_labels=True, 
    node_size=4500,
    edge_font_size=7,
    node_colours = node_col,
    show_descendant_samples="tips",
    label_replace={"Unknown (R)": "Rec node"},
    colour_metadata_key="Imputed_Nextclade_pango",
)
# Get the position array
G, pos = sc2ts.utils.plot_subgraph(list(to_plot), ts_long, ti_long, ax=ax, **params)
ax.clear()
# adjust some positions (found by trial-and-error)
pos[179] = (pos[179][0] + 90, pos[179][1]) # shift to make room for mutation labels
pos[5184] = (pos[5184][0] + 80, pos[5184][1]) # shift
pos[21656] = (pos[21656][0] + 10, pos[21656][1]) # shift
pos[5868] = (pos[5868][0] + 80, pos[5868][1]) # shift
pos[205995] = (pos[205995][0] + 70, pos[205995][1]) # shift
pos[274163] = (pos[274163][0] + 100, pos[274163][1] - 10) # shift
pos[294002] = (pos[294002][0] + 100, pos[294002][1]) # shift
pos[21605] = (pos[21605][0] - 60, pos[21605][1] + 85) # shift
pos[357852] = (pos[357852][0] + 140, pos[357852][1]) # shift
pos[240637] = (pos[240637][0] - 70, pos[240637][1]) # shift
pos[204945] = (pos[204945][0] + 60, pos[204945][1]) # shift
p = pos[232088][0]
pos[232088] = (pos[317538][0], pos[232088][1]) # shift
pos[317538] = (pos[209245][0], pos[317538][1]) # shift
pos[209245] = (pos[315067][0], pos[209245][1]) # shift
pos[315067] = (p, pos[315067][1]) # shift
pos[235293] = (pos[235293][0] - 20, pos[235293][1]) # shift
sc2ts.utils.plot_subgraph(list(to_plot), ts_long, ti_long, node_positions=pos, ax=ax, **params)
plt.show()
```

# Fraction of samples with Ns

```python
contains_ambiguous = np.zeros(ts_long.num_nodes, dtype=bool)
just_bases = np.zeros(ts_long.num_nodes, dtype=bool)
contains_gaps = np.zeros(ts_long.num_nodes, dtype=bool)

for u in ts_long.samples():
    md = ti_long.nodes_metadata[u]
    obc = md["sc2ts_qc"]["original_base_composition"]
    x = set(obc.keys())
    if x == {"A", "C", "G", "T"}:
        just_bases[u] = 1
    elif x == {"A", "C", "G", "T", "-"}:
        contains_gaps[u] = 1
    else: 
        contains_ambiguous[u] = 1
        
```

```python
np.sum(just_bases) / ts_long.num_samples
```

```python
np.sum(contains_gaps) / ts_long.num_samples
```

```python
np.sum(contains_ambiguous) / ts_long.num_samples
```
